# Chargement des données du cloud

```python
import pandas as pd
import os
import s3fs

# ================================
# 1. CHARGEMENT DES DONNÉES
# ================================

from loading import load_all_datasets

PATHS = {
    'irve': 'https://www.data.gouv.fr/api/1/datasets/r/eb76d20a-8501-400e-b336-d85724de5435',
    'revenu': 'https://www.data.gouv.fr/api/1/datasets/r/516130bc-4dcb-47f5-8347-ae96553c43ab',
    've': 'https://www.data.gouv.fr/api/1/datasets/r/90e0d717-deda-4bdc-9987-f82faac5bc93',
}

df_irve_raw, df_revenu_raw, df_ve_raw = load_all_datasets(PATHS)

print("Données chargées")

```

```python
# ================================
# 2. CONNEXION S3 (ONYXIA)
# ================================

S3_ENDPOINT_URL = "https://" + os.environ["AWS_S3_ENDPOINT"]
fs = s3fs.S3FileSystem(client_kwargs={"endpoint_url": S3_ENDPOINT_URL})

MY_BUCKET = "mtropini"  

# Création du dossier diffusion/raw
fs.mkdirs(f"{MY_BUCKET}/diffusion/raw", exist_ok=True)

```

```python
# ================================
# 3. SAUVEGARDE EN CSV
# ================================

print("Sauvegarde en cours...")

with fs.open(f"{MY_BUCKET}/diffusion/raw/irve_raw.csv", "w") as f:
    df_irve_raw.to_csv(f, index=False)

with fs.open(f"{MY_BUCKET}/diffusion/raw/revenu_raw.csv", "w") as f:
    df_revenu_raw.to_csv(f, index=False)

with fs.open(f"{MY_BUCKET}/diffusion/raw/ve_raw.csv", "w") as f:
    df_ve_raw.to_csv(f, index=False)

print("Sauvegarde terminée")

```

```python
# ================================
# 4. VÉRIFICATION DES FICHIERS
# ================================

print("Fichiers présents dans le bucket :")
print(fs.ls(f"{MY_BUCKET}/diffusion/raw"))

```

```python
# ================================
# 5. RECHARGEMENT DES DONNÉES
# ================================

print("Rechargement des données...")

with fs.open(f"{MY_BUCKET}/diffusion/raw/irve_raw.csv", "rb") as f:
    df_irve_reload = pd.read_csv(f)

with fs.open(f"{MY_BUCKET}/diffusion/raw/revenu_raw.csv", "rb") as f:
    df_revenu_reload = pd.read_csv(f)

with fs.open(f"{MY_BUCKET}/diffusion/raw/ve_raw.csv", "rb") as f:
    df_ve_reload = pd.read_csv(f)

print("Rechargement OK")

```

```python
# ================================
# 6. VÉRIFICATION FINALE
# ================================

print("Shapes rechargés :")
print("IRVE :", df_irve_reload.shape)
print("Revenu :", df_revenu_reload.shape)
print("VE :", df_ve_reload.shape)
```
